```python
import csv 
import matplotlib.pyplot as plt
def rPacks(pwdo): ##lee y ordena los paquetes no se encarga del origen y destino, ese 
    try:
        #
        with open(pwdo,'r', newline='') as entrada:
            lector = csv.reader(entrada, delimiter=',')
            listax = []
            listay = []
            for row in lector:
                listax.append(float(row[0]))
                listay.append(float(row[1]))
            return (listax,listay)
    except:
        print("Esta mal el pwd: ",pwdo)
        raise
    

def main():
    titulos = ["Mejor","Peor","Promedio"]
    pwdSerial = ["Mejor.txt","Peor.txt","Promedio.txt"]
    pwdParalelo = ["MejorP.txt","PeorP.txt","PromedioP.txt"]
    
    
    listayS = []
    listaxP =[]
    listayP = []
    (listaxS,listayS)=rPacks(pwdSerial[0]) 
    (listaxP,listayP)=rPacks(pwdParalelo[0])
    ##plt.subplot(321)
    titulo = titulos[0]
    fig = plt.figure(figsize =(20,10))
    ax = fig.add_subplot(111)
    ax.plot(listaxS,listayS,'k')
    ax.plot(listaxS,listayP,'g')##paralelo
    plt.title(titulo)
    #plt.plot()
    plt.show()
    
    
    
    
    
    listaxS = []
    listayS = []
    listaxP =[]
    listayP = []
    (listaxS,listayS)=rPacks(pwdSerial[1]) 
    (listaxP,listayP)=rPacks(pwdParalelo[1])
    titulo = titulos[1]
    fig = plt.figure(figsize =(20,10))
    ax = fig.add_subplot(111)
    ax.plot(listaxS,listayS,'k')
    ax.plot(listaxS,listayP,'g')##paralelo
    plt.title(titulo)
    #plt.plot()
    plt.show()
    
    listaxS = []
    listayS = []
    listaxP =[]
    listayP = []
    (listaxS,listayS)=rPacks(pwdSerial[2]) 
    (listaxP,listayP)=rPacks(pwdParalelo[2])
    titulo = titulos[2]
    fig = plt.figure(figsize =(20,10))
    ax = fig.add_subplot(111)
    ax.plot(listaxS,listayS,'k')
    ax.plot(listaxS,listayP,'g')##paralelo
    plt.title(titulo)
    #plt.plot()
    plt.show()         
main()

##Todo hardcodeado :'c
```
